```python
# Carregando os pacotes
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv("https://raw.githubusercontent.com/dinopolo/Datasets/master/albuns_vendidos.csv")
df_null = pd.read_csv("https://raw.githubusercontent.com/dinopolo/Datasets/master/albuns_vendidos_com_null.csv")
```

```python
sns.set(style='white')
```

```python
df.style.set_properties(**{'background-color': 'lightgrey', 'color': 'blue'})
```

```python
df.style.highlight_min(axis=0)
```

```python
df.style.highlight_max(axis=0)
```

```python
df_null.style.highlight_null()
```

```python
cm = sns.light_palette("blue", as_cmap=True)
df.style.background_gradient(cmap=cm)
```

```python
df.style.background_gradient(cmap=cm).highlight_min(axis=0, color='springgreen').highlight_max(axis=0, color='darkolivegreen')
```

```python
dfcorr = df.corr()

plt.figure(figsize=(10,8))
sns.heatmap(dfcorr, annot=True)
plt.tight_layout()
```

```python
plt.figure(figsize=(10,8))

sns.scatterplot(x="copias_vendidas_certificadas", y="copias_vendidas_estimadas", data=df, s=100)

plt.tight_layout();
```

```python
plt.figure(figsize=(10,8))

sns.countplot(data=df, x='copias_vendidas_estimadas', color='olive')

plt.tight_layout();
```
