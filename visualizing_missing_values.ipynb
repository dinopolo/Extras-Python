# **Visualizing Missing Values**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import missingno as msno

df = pd.read_csv("https://raw.githubusercontent.com/dinopolo/Datasets/master/netflix_titles.csv")
```

## **Seaborn**

```python
sns.heatmap(df.isnull(), cbar = False);
```

## **Missingno**

```python
msno.matrix(df);
```

```python
msno.bar(df);
```

```python
msno.heatmap(df);
```
