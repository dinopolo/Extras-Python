# **Treemap with Python**

## **Data Preprocessing**

```python
import pandas as pd
import squarify
import matplotlib
import matplotlib.pyplot as plt

df = pd.read_csv("https://raw.githubusercontent.com/dinopolo/Datasets/master/act_idade.csv", encoding="latin-1")
```

```python
df.Idade.value_counts()
```

```python
df = df[df.Idade != 'Ignorada']
df = df[df.Idade != '-']
```

```python
df.Idade.value_counts()
```

## **Create a Treemap Chart**

```python
squarify.plot(sizes=df.Idade.value_counts());
```

```python
df.Idade.value_counts().index
```

```python
# Definindo a paleta de cores para o treemap
cmap = matplotlib.cm.Blues
mini = min(df.Idade.value_counts())
maxi = max(df.Idade.value_counts())
norm = matplotlib.colors.Normalize(vmin=mini, vmax=maxi)
colors = [cmap(norm(i)) for i in df.Idade.value_counts()]
```

```python
# Definindo o tamanho da figura
plt.figure(figsize=(20, 12))

# Fazendo o Treemap
squarify.plot(sizes=df.Idade.value_counts(), label=df.Idade.value_counts().index, color=colors);

# Adicionando um título
plt.title('Comparativo da quantidade de acidentes de trabalho por intervalo de idade', fontsize=20)

# Removendo os eixos
plt.axis('off');

# Ajustando o gráfico
plt.tight_layout()
```

```python
# Definindo o tamanho da figura
plt.figure(figsize=(20, 12))

# Fazendo o Treemap
squarify.plot(sizes=df.Idade.value_counts(), label=df.Idade.value_counts().index, color=colors, pad=True, alpha=0.6, text_kwargs={'fontsize': 12});

# Adicionando um título
plt.title('Comparativo da quantidade de acidentes de trabalho por intervalo de idade', fontsize=20)

# Removendo os eixos
plt.axis('off');

# Ajustando o gráfico
plt.tight_layout()
```
